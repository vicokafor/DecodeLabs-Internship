### Whitespace Check

```python
text_cols = df.select_dtypes(include='object').columns

for col in text_cols:
    count = (df[col] != df[col].str.strip()).sum()
    print(f"{col}: {count}")
```

###Finding:
No white space found.

```python
df[
    round(df['Quantity'] * df['UnitPrice'], 2)
    != round(df['TotalPrice'], 2)
]
```
